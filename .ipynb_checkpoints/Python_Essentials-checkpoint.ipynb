# How did we reach to OOPs

What is the evolution of Porgrammig languages and why is OOPs so popular a paradigm today?
Coding initially was done in 0s and 1s called low level assembly languages. It was later mapped to certain keyords but these are still 
quite low and nowhere close to how humans commmunciate. 
FORTRAN and COBOl bridge those gaps and were first high level langauges. 
But as the code base grew the need to find better ways to handle code was realised- Simula 67 was the first OOPs developed in OSLO, NORWAY.
But before OOPS, the paradigm was writing functions that operate on variables/data. 
    
Imagine large teams writing functionn -
1. All operating on same variables. Changing one thing broke five others.
2. Changing one thing changes everything. With scale codebase was hard to maintain. 
3. Debugging was a nightmare.
4. Tight coupling of Data and logic in procedural coding meant changing a data structure would mean changing it at all palces. Rigidity of code. 
5. Procedural code was based on procedures operating on data which was unlike we think of real world which is like objects. 

# OOPs?


The 4 pillars work together as a system:
1. Encapsulation stops other code from messing with your internals. Data/varibables and functions were wrapped inside a class. 
Only these methods can operate on the variables of this class
2. Abstraction lets you ignore complexity you don't need to understand. 
3. Inheritance gives you reuse without duplication. A child class inherits properties and methods from a parent. 
Model "is-a" relationships naturally. A Dog is an Animal. A SavingsAccount is a BankAccount. Reuse code across hierarchies without duplication.
4. Polymorphism lets systems be extensible without being rewritten. one interface, many forms. 

Analogy
Think of an ATM machine. You cannot open the machine and change the balance. You must use controlled operations:
withdraw
deposit
check balance

```python
Q - How to remove the prefix
links = [
    "www.b001.io",
    "www.youtube.com",
    "www.google.com",
    "www.wikipedia.org"
]

for link in links:
    print(link.lstrip("www."))

Output:
b001.io
youtube.com
google.com
google.com
ikipedia.org

We can see that in the 4rth website the ‘w’ is also removed because the strip function not only removes “www.” but all combinations of it present in prefixes. Also, had we given ‘w.’ as the argument to lstrip it would have provided the same result. 

To handle this we use **`removeprefix()`**
**LEARNINGS:**

**Python String removeprefix() function** removes the prefix and returns the rest of the string. If the prefix string is not found, then it returns the original string.
    s = "HelloWorld"
res = s.removeprefix("Hello")
print(res) 
Output: World 

# No Prefix Match
s = "PythonProgramming"
res = s.removeprefix("Java")
print(res)# 
Output: PythonProgramming 

# Case Sensitivity
s = "PythonProgramming"
res = s.removeprefix("python")
print(res)# Output: PythonProgramming
```

```python
- Taking input and executing
    
    Q- 
    
    Consider a list (`list = []`). You can perform the following commands:
    
    1. `insert i e`: Insert integer  at position .
    2. `print`: Print the list.
    3. `remove e`: Delete the first occurrence of integer .
    4. `append e`: Insert integer  at the end of the list.
    5. `sort`: Sort the list.
    6. `pop`: Pop the last element from the list.
    7. `reverse`: Reverse the list.
    
    Initialize your list and read in the value of  followed by  lines of commands where each command will be of the  types listed above. Iterate through each command in order and perform the corresponding operation on your list.
    
    **Example**
    
    - : Append  to the list, .
    - : Append  to the list, .
    - : Insert  at index , .
    - : Print the array.Output:
[1, 3, 2]

**Input Format**

The first line contains an integer, , denoting the number of commands.

Each line  of the  subsequent lines contains one of the commands described above.

**Constraints**

- The elements added to the list must be *integers*.

**Output Format**

For each command of type `print`, print the list on a new line.

**Sample Input 0**

`12
insert 0 5
insert 1 10
insert 0 6
print
remove 6
append 9
append 1
sort
print
pop
reverse
print`

**Sample Output 0**

`[6, 5, 10]
[1, 5, 9, 10]
[9, 5, 1]`

My Solution:
        N = int(input())
    nlist=list()
    for i in range(N):
        command=input()
        if command.split()[0]=='insert':
            nlist.insert(int(command.split()[1]),int(command.split()[2]))
        if command.split()[0]=='print':
            print(nlist)
        if command.split()[0]=='remove':
            nlist.remove(int(command.split()[1]))
        if command.split()[0]=='append':
            nlist.append(int(command.split()[1]))
        if command.split()[0]=='sort':
            nlist.sort()
        if command.split()[0]=='pop':
            nlist.pop()
        if command.split()[0]=='reverse':
            nlist.reverse()
Improved solution:

N = int(input())
clist=rlist=list()
for i in range(N):
    command,*line=input().split()  # A way to collect all similar types of input together. SPLIT() outputs the results as a list().
    nline=list(map(int,line))  # Using Map fucntion, converting the similar type collection to one datatype and then storing it into a list object.
    if command=='append':
        rlist.append(nline[0])
    elif command=='insert':
        rlist.insert(nline[0],nline[1])
    elif command=='remove':
        rlist.remove(nline[0])
    elif command=='print':
        print(rlist)
    elif command=='sort':
        rlist.sort()
    elif command=='pop':
        rlist.pop()
    elif command=='reverse':
        rlist.reverse()
```

```python
sWaP CaSe
You are given a string and your task is to *swap cases*. In other words, convert all lowercase letters to uppercase letters and vice versa.

**For Example:**
Www.HackerRank.com → wWW.hACKERrANK.COM
Pythonist 2 → pYTHONIST 2

def swap_case(s):
    temp=list(s)
    final=list()
    for i in temp:
        if i.isalpha() and i.islower():
            final.append(i.upper())
        elif i.isalpha() and i.isupper():
            final.append(i.lower())
        else:
            final.append(i)
    
    return ''.join(map(str,final))

if __name__ == '__main__':
    s = input()
    result = swap_case(s)
    print(result)

    **LEARNINGS:**

1. Strings have an attribute **swapcase()** that can be directly used as - `str1.swapcase()`
2. we can convert a string to a list of individual alphabets using - `list(str)`
3. We can create a list from strings directly using **split()** as well

str1="Hel_lo"
temp=str1.split("_")
print(type(temp))

output- <class 'list'>

4. We can join back a list of strings or mixed datatypes as strings using JOIN().
    1. For a list of strings-
a = ['Geeks', 'for', 'Geeks']

# Convert list to string using a loopres = ''
for sin a:
    res += s + ' '

# Remove trailing spaceres = res.strip()
print(res)


or 

res = ' '.join(a)
print(res)

For a mixed data types of list we can either go for list comprehension or use MAP function

List comprehension

a = [1, 'apple', 3.14, 'banana']
res = ' '.join([str(s)for sin a])
print(res)

or MAP()

res = ' '.join(map(str, a))
```

```python
Counting Substrings in a string
def count_substring(string, sub_string):
    counter=0
    for i in range(len(string)):
        if string[i:i+len(sub_string)]==sub_string:
            counter+=1
    return counter

if __name__ == '__main__':
    string = input().strip()
    sub_string = input().strip()
    
    count = count_substring(string, sub_string)
    print(count)
    
 or
 print(sum([ 1 for i in range(len(string)-len(substring)+1) if string[i:i+len(substring)] == substring]))

Improved Solution:

def count_substring(string, sub_string,count=0):
    for i in range(len(string)):
        if string.find(sub_string,i,i+len(sub_string))!=-1:
            count+=1
    return count


    **LEARNINGS:**

1. String.startswith() or string.find() method is used to find the occurance of a string inside a string.
    1. `string.find(sub_string)` will return the index of the first occurance.
    2. `*string*.find(*value, start, end*)` will customise the search

    start	Optional. Where to start the search. Default is 0
    end	Optional. Where to end the search. Default is to the end of the string
```

```python
Check for alnumeric, alphabets, digits, upper, lower cases in a string

    s=input()
    c_alnum=c_alpha=c_d=c_l=c_u=0
    for i in range(len(s)):
        if s[i].isalnum():
            c_alnum+=1        
        if s[i].isalpha():
            c_alpha+=1
        if s[i].isdigit():
            c_d+=1
        if s[i].islower():
            c_l+=1
        if s[i].isupper():
            c_u+=1
    print(c_alnum>0)
    print(c_alpha>0)
    print(c_d>0)
    print(c_l>0)
    print(c_u>0)

Improved Solution
print(any(c.isalnum() for c in s))
print(any(c.isalpha() for c in s))
print(any(c.isdigit() for c in s))
print(any(c.islower() for c in s))
print(any(c.isupper() for c in s))

**LEARNINGS**

1. An **alphanumeric** string is a string that contains **only**: **Alphabets (letters)**: `A–Z` and `a–z or`**Digits**: `0–9` and **No spaces, punctuation, special characters** like `@`, `#`, `!`, `/`, `,`, etc.
2. ANY() takes an iterable as argument. (An iterable object -Strings**,**list, set, tuple, dictionary etc)
    1. The `any()` function returns True if any item in an iterable are true, otherwise it returns False.
    2. If the iterable object is empty, the `any()` function will return False.
```
