````python
Forward slash vs Backward slashes

Which of the following path will not raise an error while reading in pandas
1. `'C:\Users\Mandeep\Downloads\winequality-white'`
2. `'C:/Users/Mandeep/Downloads/winequality-white'`

In Python, both of the paths you provided are valid. However, when working with file paths in Python, it's generally recommended to use forward slashes ('/') or raw string literals (r'') to avoid potential issues with escape sequences and platform compatibility.

**Path using forward slashes ('/') is more Pythonic** and **platform-independent**. It will work on both Windows and Unix-like operating systems without modification.

**Why is forward slash necessary in python read file**

In Python, both forward slashes ('/') and backslashes ('\') can be used to specify file paths when reading files. However, there are a few reasons why forward slashes are typically preferred:

1. **Platform Independence**: Forward slashes are the standard path separator in Unix-like operating systems (e.g., Linux, macOS), while backslashes are typically used in Windows. However, Python supports both forward and backward slashes as path separators on all platforms. Using forward slashes ensures that your code remains platform-independent and can be easily executed on any operating system without modification.
2. **Escape Sequences**: Backslashes are used as escape characters in Python strings. For example, '\n' represents a newline character, '\t' represents a tab character, etc. **If you use backslashes in file paths without escaping them (e.g., 'C:\Users\Mandeep\Downloads\file.txt'), Python may interpret them as escape sequences, leading to unexpected behavior or errors. To avoid this, you can either use raw string literals (r'') or double backslashes ('\') to escape the backslashes in file paths**. However, using forward slashes eliminates the need for escaping altogether.
3. **Readability**: Forward slashes are more commonly used in Python code and are generally considered more readable than backslashes, especially for specifying file paths. Using forward slashes makes your code easier to understand and maintain, especially for developers who are more familiar with Unix-style conventions.

Overall, while both forward and backward slashes can be used in Python file paths, forward slashes are preferred for their platform independence, avoidance of escape sequences, and better readability. However, Python provides flexibility, allowing you to choose the path separator that best suits your needs and preferences.
````

````python
Essential functions and Modules

1. split() and strip() function
    1. If you want to remove leading and trailing spaces from a string, you can use the strip() method. Whereas **x.split()** splits the string **into a LIST** of substrings based on whitespace or other delimiter provided. 
    2. Why df[column].str.strip() ran but df[column].strip() does not
        
        `df[column]` is a **Pandas Series** (a column of values). 
        
        `.str` accesses **string methods** element-wise for the whole Series.
        `.strip()` removes leading/trailing spaces from **each string** in that column. `.strip()` is a method for **a single Python string**, not a Pandas Series. Does it also strip in between a string? NO
        
        Instead, it should be used on individual string elements or within a lambda function when applying it to each element in the Series. The strip() method is a string method in Python, and it is typically used on individual strings, not on entire columns in a DataFrame. **When working with a DataFrame column, you should use the str.strip() method, which applies to a series of strings.**
        
        **NOTE**: To strip only left or right side we can use text.lstrip() or text.rstrip().

2. 1. **append-concat-merge**?
    - Why use of **`np.append`** instead of **`pd.append`** while adding values to a list? **`np.append`** is a NumPy function, and it's used to concatenate or append elements to an array. The **`pd.append`** method, on the other hand, is a method associated with pandas DataFrames and Series, but it's not typically used for appending elements to an existing DataFrame in a row-wise fashion.
        
        **`pd.append`**, **`pd.concat`**, and **`pd.merge`** are three different methods in pandas, each serving a specific purpose when it comes to combining or appending DataFrames. Here's a brief overview of the differences:
        
        1. **`pd.append`**:
            - The **`pd.append`** method is used to append rows from one DataFrame to another. It concatenates rows along the vertical axis (axis=0).
            - It takes another DataFrame as an argument and appends its rows to the original DataFrame.
            - It returns a new DataFrame and does not modify the original DataFrames in place.
            - It is simple and useful for appending a small number of rows.
            
            **`pd.concat`**:
            
            - The **`pd.concat`** function is more flexible and can concatenate DataFrames along either axis (rows or columns).
            - It can concatenate multiple data frames along a specified axis, handling different shapes and indices.
            - It returns a new DataFrame and does not modify the original DataFrames by default.
            - **Simple Stacking of DataFrames**: It is useful when you want to concatenate DataFrames along an existing axis **without performing any matching** based on columns. That is **no Requirement for Common Columns**: concat does not require the DataFrames to have the same columns. It simply stacks the DataFrames along the specified axis.
            - It is suitable for combining larger datasets and handling various concatenation scenarios.
            **Concatenation along an Axis**: concat is primarily used for concatenating two or more data frames along a particular axis, either rows (axis=0) or columns (axis=1).
        **`pd.merge`:**

1. **Database-style Merging:**
    - **`merge`** is used for combining DataFrames based on the values of common columns. It performs operations similar to SQL joins.
2. **Joining on Specific Columns:**
    - It requires the specification of the columns on which to join the DataFrames (**`on`** parameter).
3. **Combining Based on Key Columns:**
    - **`merge`** is typically used when you want to combine DataFrames based on shared columns (keys), and it provides various options for the type of join (e.g., inner, outer, left, right).
4. It performs relational algebra operations similar to SQL joins, allowing you to merge DataFrames based on specified keys.
5. It returns a new DataFrame resulting from the merge operation.
    - It is ideal for combining DataFrames when there is a common column or index on which to merge.

    In summary:

- Use **`pd.append`** for simple row-wise concatenation.
- Use **`pd.concat`** for more flexible concatenation along either axis. It is used for simple stacking of DataFrames along an axis without the need for column-wise matching.
- Use **`pd.merge`** for merging DataFrames based on common columns or indices.

3. get()
1. In pandas, the **`get()`** method is used to access a value for a given key. Specifically, it is designed to retrieve the value associated with a specified key in a pandas DataFrame or Series. This method is useful when you want to access a value based on its key, similar to how you might access elements in a dictionary.
- Dictionaries
    - Diff in dict.get(key) and dict[key]?
        
        The difference between **`dict.get(key)`** and **`dict[key]`** lies in how they behave when the specified key is not found in the dictionary
        
        1. **`dict.get(key)`**:
            - This method returns the value associated with the specified key in the dictionary.
            - If the key is found in the dictionary, it returns the corresponding value.
            - If the key is not found in the dictionary, it returns **`None`** by default (or a specified default value if provided as the second argument).
    2. **`dict[key]`**:
    - This syntax retrieves the value associated with the specified key in the dictionary.
    - If the key is found in the dictionary, it returns the corresponding value.
    - If the key is not found in the dictionary, it raises a **`KeyError`** exception.
    In summary:

- Use **`dict.get(key)`** when you want to retrieve a value from the dictionary and handle the case where the key may not exist without raising an error.
- Use **`dict[key]`** when you are confident that the key exists in the dictionary and you want the code to raise an error if the key is not found.
````

````python
- What is seed in random?
    
    In the context of random number generation, a seed is a starting point used to initialize a pseudorandom number generator (PRNG). Pseudorandom number generators are algorithms that produce sequences of numbers that appear random but are actually determined by an initial value called the seed.
    
    Here's how the seed works:
    
    1. **Initialization:** When you set a seed value, you're initializing the pseudorandom number generator to a specific starting state. This starting state will determine the sequence of random numbers that the generator produces.
    2. **Reproducibility:** By setting the seed to a specific value, you ensure that the sequence of random numbers generated by the PRNG is reproducible. In other words, if you use the same seed value in different runs of your program, you'll get the same sequence of random numbers every time.
    3. **Randomness:** Despite being deterministic (i.e., producing the same sequence given the same seed), PRNGs are designed to produce sequences of numbers that exhibit properties of randomness. This means that the numbers in the sequence should be statistically indistinguishable from truly random numbers.
    
    In Python, you can set the seed using the **`random.seed()`** function from the **`random`** module

By setting the seed to the same value (in this case, 42), you ensure that random_number will always be the same, regardless of how many times you run the program. This can be useful for debugging, testing, and ensuring reproducibility in your code.
````

````python
- Regular expressions
    - What are different functions in re module and what they returns
        
        The **`re`** module in Python provides several functions for working with regular expressions. Here's an overview of some of the key functions along with their descriptions and return values:
        
        1. **`re.search(pattern, string)`**:
            - Description: Searches for the first occurrence of a pattern within a string.
            - Returns: A match object if the pattern is found, or **`None`** if the pattern is not found.
        2. **`re.match(pattern, string)`**:
            - Description: Checks if the pattern matches at the beginning of the string.
            - Returns: A match object if the pattern matches at the beginning of the string, or **`None`** otherwise.
        3. **`re.findall(pattern, string)`**:
            - Description: Finds all occurrences of a pattern within a string.
            - Returns: A list of all non-overlapping matches as strings.
        4. **`re.finditer(pattern, string)`**:
            - Description: Finds all occurrences of a pattern within a string.
            - Returns: An iterator yielding match objects for each match found.
        5. **`re.sub(pattern, replacement, string)`**:
            - Description: Replaces occurrences of a pattern within a string with a specified replacement.
            - Returns: A new string with all occurrences of the pattern replaced by the replacement.
        6. **`re.split(pattern, string)`**:
            - Description: Splits a string into a list of substrings based on a pattern.
            - Returns: A list of substrings.
        7. **`re.compile(pattern)`**:
            - Description: Compiles a regular expression pattern into a regular expression object, which can be used for matching, searching, etc.
            - Returns: A compiled regular expression object.
        8. **`re.fullmatch(pattern, string)`**:
            - Description: Checks if the entire string matches the pattern.
            - Returns: A match object if the entire string matches the pattern, or **`None`** otherwise.
        9. **`re.escape(string)`**:
            - Description: Escapes special characters in a string so that they can be used as literals in a regular expression.
            - Returns: A new string with special characters escaped.
        
        These are some of the most commonly used functions in the **`re`** module. They provide powerful tools for working with regular expressions in Python, allowing you to search, match, replace, and manipulate strings based on complex patterns.
````

````python
- Datetime
    
     `datetime` module is packed with useful methods for working with dates and times.
    
    - Basics
        
        `from datetime import datetime, date, time`
        
        `datetime.now()`      # Current date & time (with time)
        `date.today()`        # Current date only
        `datetime.utcnow()`   # Current UTC date & time
        
        `date(2025, 8, 8)`                   # Create a date (YYYY, M, D)
        `time(14, 30, 45)`                   # Create a time (HH, MM, SS)
        `datetime(2025, 8, 8, 14, 30, 45)`   # Create full datetime
        
    - Formatting Dates & Times
        
        `now = datetime.now()`
        
        `now.strftime("%Y-%m-%d %H:%M:%S")`  # → '2025-08-08 11:45:32'
        `now.strftime("%A, %d %B %Y")`       # → 'Friday, 08 August 2025'
        Common format codes:
        
        %Y → 4-digit year
        
        %m → 2-digit month
        
        %d → Day of month
        
        %H → Hour (24-hour)
        
        %M → Minute
        
        %S → Second
        
        %A → Full weekday name
        
        %B → Full month name
        
        NOTE: For converting a datetime to a string we can use the belwo 2 methods:
        
        1. Use `str(datetime_variable)`. This just do the direct conversion ad not customisable.
        2. Use `datetime_variable.strftime(”%Y-%m-%d %H:%M:%S”)` . Usinf strftime we can customise each part of the datetime. 
    - Parsing Strings into Dates
        
        `datetime.strptime("2025-08-08", "%Y-%m-%d")`
        
    - Extracting Parts of a Date
        
        now = datetime.now()
        
        now.year      # 2025
        now.month     # 8
        now.day       # 8
        now.hour      # 11
        now.minute    # 45
        now.second    # 32
        
    - Doing Date Arithmetic
        
        `from datetime import timedelta`
        
        `today = date.today()
        yesterday = today - timedelta(days=1)
        tomorrow = today + timedelta(days=1)`
        
        `delta = tomorrow - today`   # → timedelta(days=1)
        `delta.days`                 # → 1
        
    - Weekday Methods
        
        `today = date.today()`
        
        `today.weekday()`  # Mon=0, Sun=6
        `today.isoweekday()`  # Mon=1, Sun=7
````

````python
- Tidy data
    
    What is tidy Data?
    
    []()
    
    I was trying to transpose this data which I did easily with transpose shortcut but when I tried this with Pivot, I was not able to.
    
    The reason was simple that this above data doesnot follow TIDY Data principles. Pivot Tables Work on “Long” (Tidy) Format Data.
    
    Tidying data is the process of, “[structuring datasets to facilitate analysis.](https://cran.r-project.org/web/packages/tidyr/vignettes/tidy-data.html)“
    
    Every dataset you analyze is going to come to you structured in a unique way, so the goal is to reformat the data to a standard format every single time. Standardizing datasets in this way makes it easier for you to analyze each dataset, but it also makes it easier for people to work together or on teams.
    
    Tidy data works on three core principles:
    
    1. Each variable must have its own column
    2. Each observation must have its own row
    3. Each value must have its own cell
    
    In our dataset, it is already a pivot and we need to unpivot/tidy it before we can pivot it again in other ways.
    
    In this wide format:
    
    - **Days are stored as columns**, which Pivot Tables don’t directly use for rows.
    - You can’t tell Pivot Table: “Make these column headers into row values” — it doesn’t know how.
    
    According to tidy data principles each Day should not have its own column. Instead, they should be gather together into a Day/Date column.
    
    Does it means if any table is already pivoted then cannot be pivoted it again?
    
    Pivot Tables work best on **raw, long/flat, tidy** data. If a table is already in a "pivoted" or "summary" form (wide layout), it's no longer raw data. It’s **aggregated**, and the structure **no longer tells Excel how to re-summarize it**.
````
