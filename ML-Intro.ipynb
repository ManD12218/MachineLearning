What do you understand by ML? Why do we use ML?
1. ML is about making machines get better at some task by learning from the data instead of having to explicitly code. 
    1. Why? Some tasks cannot be defined well, except by examples.
    – It is very hard to write programs that solve problems like recognizing a handwritten digit
    – What distinguishes a 2 from a 7?
    How does our brain do it
    • Hidden relationships and correlations in data
    • large data makes it difficult for explicit encoding by
    humans (e.g., medical diagnostic)
    • Continuous availability of new knowledge
    
    That is, a computer is said to learn from experience E with respect to some task T, and some performance measure P, if its performance on T, measured by P, improves with experience E.
    

1. Why do we use ML? 
    1. Dynamic problems where existing rules become obscure readily and new rules need to be written every time. 
    2. To get insights from complex problems and large amounts of data.
2. Examples of tasks achieved by using ML
    1. CLASSIFICATION- Analysing images of products on a production line to automatically classify them- uses CNN (Convolutional Neural Networks)
    2. ANAMOLY DETECTION- Detecting tumors in Brain scans (CNN), Detecting Credit card fraud
    3. CLUSTERING- Mail Spam filter, segmenting clusters in customers for marketing campaigns
    4. VISUALISATION- Visualizing high dimensional dataset using dimensionality reduction
    5. NATURAL LANGUAGE PROCESSING- speech recognition in apps, and chatbots using NLU (NL Understanding)
    6. FORECASTING- regression for forecasting revenues of a business.

```python
Types of ML Algorithms

The types depend upon whether:
1. The system is trained with human supervision
1.1 Does the training data already consists of solutions with labels.
Yes then Supervised - Classification, Regression tasks
No then UnSupervised - Clustering, Anomaly detection, Association Rule Learning, Dimensionality reduction
Few then SemiSupervised. Google photos use this

2. Can the system learn incrementally on the fly
Batch learning - The system is trained offline and then launched into production. Takes a lot of time and computing resources.
Incremental / Online learning - The system is trained incrementally by feeding new data in mini-batches making learning fast and cheap, 
while the system is online. 

3. How does the system deal with new instances in the data
Instance-based - Compare it with old data. With new data, it uses a similarity measure to identify patterns. Ex- Spam filtering
Model-based - Find patterns in training data. A model that best fits the training data is used to generalize (or forecast). Ex- Linear regression.


Reinforced learning
This is different. the learning system is called an agent, which can observe an environment, select and perform an action, and get rewarded or 
penalized. The system has to learn itself the best strategy called policy by collecting the most rewards.
```

```python
Challenges of ML

1. Sufficiency of Data- ML requires a lot of data for solving accurately.
2. Non-representative Training Data- If the data set is small then it would have high sampling noise. But even if the data set is large, it can have sample Bias. 
3. Poor Quality Data- Data full of errors, outliers, and noise will require a lot of data cleaning.
4. Irrelevant features- Too many irrelevant features require 
    1. feature selection- Selecting the most useful features to train data.
    2. feature extraction (combining existing features to produce more useful features)
5. Overfitting the training data- means models perform well on training data but it does not generalize well. This happens when the model is too complex relative to the amount and noise in training data. The possible solutions are:
    1. Simplify the model by selecting fewer parameters.
        1. Constraining a model to make it simpler and reduce the risk of overfitting is called regularisation. The amount of regularisation to apply can be controlled by a Hyperparameter. 
    2. Gather more training data
    3. Reduce noise (Fix data errors and remove outliers)
6. Underfitting the training data- When a model is too simple to understand its underlying structure. The possible solutions are:
    1. Select a more powerful model with more parameters
    2. Feature engineering- feed better features 
    3. Reduce the constraints on the model- reduce the regularisation hyperparameter.
```
